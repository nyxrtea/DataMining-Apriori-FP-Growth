# **Penerapan Algoritma**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/datmin.csv')
df_numeric = df.select_dtypes(include=['int64', 'float64'])

if 'OUTPUT Grade' not in df_numeric.columns:
    df_numeric['OUTPUT Grade'] = df['OUTPUT Grade'].astype('category').cat.codes  # encode grade

correlation_matrix = df_numeric.corr()
plt.figure(figsize=(14,10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap Korelasi Fitur terhadap OUTPUT Grade")
plt.show()

korelasi_target = correlation_matrix["OUTPUT Grade"].abs()
fitur_terpilih = korelasi_target[korelasi_target > 0.2].index.tolist()
fitur_terpilih.remove("OUTPUT Grade")

print("Fitur yang dipilih berdasarkan korelasi:")
print(fitur_terpilih)
```

Syntax ini bertujuan untuk melakukan analisis korelasi antara berbagai fitur numerik dalam dataset dengan variabel target bernama OUTPUT Grade.

Setelah itu, kode menghitung korelasi antar fitur numerik, termasuk dengan kolom target, menggunakan korelasi Pearson yang mengukur kekuatan dan arah hubungan linier antar variabel. Hasil korelasi ini kemudian divisualisasikan dalam bentuk heatmap, yaitu peta warna yang memperlihatkan seberapa kuat dan positif atau negatif hubungan setiap fitur dengan fitur lain, khususnya dengan OUTPUT Grade.

Dari hasil korelasi tersebut, kode memilih fitur-fitur yang memiliki korelasi absolut lebih besar dari 0.2 terhadap OUTPUT Grade, artinya fitur-fitur tersebut memiliki hubungan yang cukup signifikan baik positif maupun negatif dengan nilai akhir. Fitur-fitur ini dipandang sebagai variabel penting yang berpengaruh dan layak untuk dipakai dalam analisis lanjutan, seperti membuat model prediksi.

Secara keseluruhan, kode ini digunakan untuk mengidentifikasi dan memvisualisasikan fitur-fitur yang paling relevan dengan nilai akhir dalam dataset, sehingga membantu dalam pengambilan keputusan terkait pemilihan variabel yang akan digunakan dalam proses analisis atau pemodelan data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python

df_subset = df[fitur_terpilih + ['OUTPUT Grade']]
transactions = df_subset.astype(str).apply(lambda row: [f"{col}:{val}" for col, val in row.items()], axis=1).tolist()
transactions = [[str(item) for item in transaction] for transaction in transactions]
```

```python

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

```python

supports = [0.05, 0.1, 0.15, 0.2]
confidences = [0.6, 0.7, 0.8, 0.9]

results = []
for support in supports:
    for confidence in confidences:
        itemsets = apriori(df_encoded, min_support=support, use_colnames=True)
        if not itemsets.empty:
            rules = association_rules(itemsets, metric="confidence", min_threshold=confidence)
            jumlah_aturan = len(rules)
        else:
            jumlah_aturan = 0
        results.append([support, confidence, jumlah_aturan])
```

```python

results_df = pd.DataFrame(results, columns=["support", "confidence", "jumlah_aturan"])
pivot_table = results_df.pivot(index="support", columns="confidence", values="jumlah_aturan")

# Plot heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Jumlah Aturan berdasarkan Support dan Confidence")
plt.xlabel("Confidence")
plt.ylabel("Support")
plt.tight_layout()
plt.show()
```

Gambar heatmap di atas menunjukkan jumlah aturan asosiasi yang dihasilkan berdasarkan kombinasi nilai support dan confidence. Sumbu vertikal (baris) menunjukkan variasi nilai support, yaitu 0.05, 0.1, 0.15, dan 0.2, sedangkan sumbu horizontal (kolom) menunjukkan nilai confidence, yaitu 0.6, 0.7, 0.8, dan 0.9. Setiap sel dalam heatmap berisi angka yang menunjukkan banyaknya aturan asosiasi yang ditemukan untuk kombinasi nilai support dan confidence tertentu. Misalnya, pada kombinasi support = 0.05 dan confidence = 0.6, ditemukan sebanyak 122 aturan.

Dari pola warna yang ditampilkan, terlihat bahwa semakin rendah nilai support dan confidence, semakin banyak aturan yang dihasilkan. Hal ini tergambar dari warna biru gelap yang menunjukkan jumlah aturan yang tinggi, sedangkan warna kuning terang menunjukkan jumlah aturan yang lebih rendah. Secara umum, jumlah aturan cenderung menurun seiring dengan meningkatnya nilai support dan confidence karena syarat untuk membentuk aturan menjadi semakin ketat.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules

def run_algorithms(df_encoded, support, confidence):
    result = {}

    # Apriori
    start_a = time.time()
    freq_apriori = apriori(df_encoded, min_support=support, use_colnames=True)
    rules_apriori = association_rules(freq_apriori, metric="confidence", min_threshold=confidence)
    end_a = time.time()

    result['Apriori_Time'] = end_a - start_a
    result['Apriori_Rules'] = len(rules_apriori)
    result['Apriori_Output'] = rules_apriori

    # FP-Growth
    start_f = time.time()
    freq_fp = fpgrowth(df_encoded, min_support=support, use_colnames=True)
    rules_fp = association_rules(freq_fp, metric="confidence", min_threshold=confidence)
    end_f = time.time()

    result['FPGrowth_Time'] = end_f - start_f
    result['FPGrowth_Rules'] = len(rules_fp)
    result['FPGrowth_Output'] = rules_fp

    return result

kombinasi = [
    {'support': 0.05, 'confidence': 0.6},
    {'support': 0.1, 'confidence': 0.8}
]
```

# **Perbandingan**

```python
for i, params in enumerate(kombinasi, 1):
    print(f"\n=== Kombinasi {i}: Support={params['support']}, Confidence={params['confidence']} ===")
    hasil = run_algorithms(df_encoded, params['support'], params['confidence'])

    print(f"Apriori - Waktu: {hasil['Apriori_Time']:.4f}s, Aturan: {hasil['Apriori_Rules']}")
    print(f"FP-Growth - Waktu: {hasil['FPGrowth_Time']:.4f}s, Aturan: {hasil['FPGrowth_Rules']}\n")

    print(">> Apriori - Aturan Teratas:")
    print(hasil['Apriori_Output'][['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())

    print("\n>> FP-Growth - Aturan Teratas:")
    print(hasil['FPGrowth_Output'][['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())
```

# **Visualisasi**

```python
import matplotlib.pyplot as plt
import seaborn as sns

hasil_list = []
for combo in kombinasi:
    result = run_algorithms(df_encoded, combo['support'], combo['confidence'])
    hasil_list.append({
        'Support': combo['support'],
        'Confidence': combo['confidence'],
        'Algoritma': 'Apriori',
        'Jumlah Aturan': result['Apriori_Rules'],
        'Waktu (detik)': result['Apriori_Time']
    })
    hasil_list.append({
        'Support': combo['support'],
        'Confidence': combo['confidence'],
        'Algoritma': 'FP-Growth',
        'Jumlah Aturan': result['FPGrowth_Rules'],
        'Waktu (detik)': result['FPGrowth_Time']
    })

df_visual = pd.DataFrame(hasil_list)

# Plot jumlah aturan
plt.figure(figsize=(10, 6))
sns.barplot(data=df_visual, x='Support', y='Jumlah Aturan', hue='Algoritma', ci=None)
plt.title('Perbandingan Jumlah Aturan Apriori vs FP-Growth')
plt.ylabel('Jumlah Aturan')
plt.xlabel('Support')
plt.legend(title='Algoritma')
plt.tight_layout()
plt.show()
```

Grafik ini dihasilkan dari proses  yang menjalankan kedua algoritma pada kombinasi nilai support dan confidence tertentu, saat algoritma dijalankan, jumlah aturan yang dihasilkan dan waktu eksekusinya disimpan dalam sebuah list, lalu dikonversi menjadi DataFrame untuk divisualisasikan.

```python
import networkx as nx

top_rules = hasil['Apriori_Output'].sort_values(by="lift", ascending=False).head(10)

G = nx.DiGraph()

for _, row in top_rules.iterrows():
    for antecedent in row['antecedents']:
        for consequent in row['consequents']:
            G.add_edge(antecedent, consequent, weight=row['lift'])

plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, k=1)
edges = G.edges(data=True)
weights = [edge[2]['weight'] for edge in edges]

nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color=weights,
        width=2.0, edge_cmap=plt.cm.viridis, arrows=True)
plt.title("Network Graph Aturan Asosiasi (Apriori)")
plt.show()
```

Gambar di atas menampilkan network graph atau graf asosiasi dari 10 aturan asosiasi teratas yang dihasilkan oleh algoritma Apriori, berdasarkan nilai lift tertinggi. Setiap node pada grafik ini mewakili sebuah item atau atribut dalam dataset, sedangkan panah mewakili aturan asosiasi yang menghubungkan sebab akaibat. Arah panah menunjukkan arah asosiasi, dari bagian kiri (antecedent) ke kanan (consequent). Ketebalan dan warna panah ditentukan oleh nilai lift; semakin tinggi nilai lift, semakin tebal warna panah tersebut, menunjukkan aturan yang lebih kuat dan lebih signifikan.

Sebagai contoh, node seperti "Sex:2" atau "Impact of your projects/activities on your success:1" berperan sebagai antecedent yang memiliki hubungan ke atribut lain, seperti "OUTPUT Grade:1" atau "Expected Cumulative grade point average in the graduation (/4.00):4". Hubungan-hubungan ini menunjukkan pola atau kecenderungan yang ditemukan dalam data, misalnya  jenis kelamin tertentu atau persepsi terhadap proyek bisa diasosiasikan dengan capaian akademik tertentu.


```python
import networkx as nx
import matplotlib.pyplot as plt

top_rules_fp = hasil['FPGrowth_Output'].sort_values(by="lift", ascending=False).head(10)
G = nx.DiGraph()

for _, row in top_rules_fp.iterrows():
    for antecedent in row['antecedents']:
        for consequent in row['consequents']:
            G.add_edge(antecedent, consequent, weight=row['lift'])
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, k=1)
edges = G.edges(data=True)
weights = [edge[2]['weight'] for edge in edges]

nx.draw(G, pos, with_labels=True, node_color='lightgreen', edge_color=weights,
        width=2.0, edge_cmap=plt.cm.viridis, arrows=True)
plt.title("Network Graph Aturan Asosiasi (FP-Growth)")
plt.show()
```

Visualisasi ini menampilkan graf jaringan aturan asosiasi berdasarkan hasil algoritma FP-Growth, yang menunjukkan hubungan antar variabel dalam dataset mahasiswa. Tujuan utama dari visualisasi ini adalah untuk memetakan hubungan asosiasi terkuat antara atribut-atribut dalam data, berdasarkan nilai lift tertinggi dari aturan yang ditemukan. Dalam konteks ini, setiap panah yang menghubungkan satu node ke node lain merepresentasikan sebuah aturan asosiasi, di mana node asal (antecedent) memiliki hubungan asosiasi terhadap node tujuan (consequent). Dari visualisasi tersebut, terdapat beberapa pola, mahasiswa dengan jenis kelamin 2 tampaknya cenderung memiliki harapan IPK kelulusan 4.00, serta berasosiasi dengan output grade sebesar 7. Selain itu, mahasiswa dengan IPK semester terakhir yang tinggi (5.00) juga menunjukkan persepsi positif terhadap dampak dari proyek atau aktivitas mereka, menunjukkan hubungan antara performa akademik dan pengalaman proyek. Harapan IPK yang tinggi juga berkorelasi dengan persepsi positif terhadap proyek, dan persepsi tersebut selanjutnya berhubungan dengan nilai output yang tinggi.

Secara keseluruhan, visualisasi ini memperlihatkan bahwa jenis kelamin, IPK sebelumnya, dan persepsi terhadap proyek memiliki peran penting dalam membentuk ekspektasi akademik serta memengaruhi hasil akhir mahasiswa.
